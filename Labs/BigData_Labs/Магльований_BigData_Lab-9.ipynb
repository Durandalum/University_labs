```python
import numpy as np
import random as r
```

```python
def matrix_by_random(size:int):
  matrix = np.array([])
  for i in range(0, size):
    for j in range(0, size):
      matrix = np.append(matrix, r.randint(-5, 15))
  matrix = matrix.reshape(size, size)
  return matrix
```

```python
def matrix_by_file():
  matrix = np.array([])
  from google.colab import drive
  drive.mount('/content/drive')
  filepath = "/content/drive/MyDrive/Colab Notebooks/BigData_L-6.txt"
  file = open(filepath, "r")
  data = file.read()
  data = data.replace("\n", " ")
  data = data.split(" ")
  for i in range(len(data)):
    matrix = np.append(matrix, int(data[i]))
  matrix = matrix.reshape(int(np.sqrt(len(data))), int(np.sqrt(len(data))))
  return matrix
```

```python
def normal_matrix(matrix:int):
  normal = 0.0
  max = 0.0
  for i in range(0, len(matrix)):
    max = 0
    for j in range(0, len(matrix)):
      max += np.absolute(matrix[i][j])
    if (max > normal):
      normal = max
  return normal
```

```python
def qualifier(matrix:int):
  a = 0.0
  b = 0.0
  c = 0
  if (len(matrix) > 2):
    for i in range(0, len(matrix)):
      b = 1.0
      for j in range(0, len(matrix)):
        c = j + i
        if (c >= len(matrix)):
          c -= len(matrix)
        b = b * matrix[j][c]
      a += b
    for i in range(len(matrix) - 1, -1, -1):
      b = 1.0
      for j in range(0, len(matrix)):
        c = i - j
        if (c < 0):
          c += len(matrix)
        b = b * matrix[j][c]
      a -= b
  else:
    a = (matrix[0][0] * matrix[1][1]) - (matrix[0][1] * matrix[1][0])
  return a
```

```python
def inverted(matrix:int):
  buffer = 0.0
  qual = qualifier(matrix)

  if (qual != 0):
    transpose = matrix
    for i in range(0, len(matrix)):
      for j in range(i + 1, len(matrix)):
        buffer = transpose[j][i]
        transpose[j][i] = transpose[i][j]
        transpose[i][j] = buffer

    minor = np.zeros((len(matrix), len(matrix)))
    auxilary = np.zeros((len(matrix) - 1, len(matrix) - 1))
    for i in range(0, len(matrix)):
      for j in range(0, len(matrix)):
        for k in range (0 , len(matrix)):
          for l in range (0, len(matrix)):
            if (k != i and l != j):
              if (k > i):
                if (l > j):
                  auxilary[k - 1][l - 1] = transpose[l][k]
                else:
                  auxilary[k - 1][l] = transpose[l][k]     
              else:
                if (l > j):
                  auxilary[k][l - 1] = transpose[l][k]
                else:
                  auxilary[k][l] = transpose[l][k]
        if (len(auxilary) >= 2):
          minor[i][j] = qualifier(auxilary)
        else:
          minor[i][j] = auxilary[0][0]
    reversed = np.zeros((len(matrix), len(matrix)))
    for i in range(0, len(matrix)):
      for j in range(0, len(matrix)):
        reversed[j][i] = (minor[j][i] * (1 / qualifier(matrix)))
  else:
    print("Определитель матрицы равняется нулю!")
  return reversed
```
