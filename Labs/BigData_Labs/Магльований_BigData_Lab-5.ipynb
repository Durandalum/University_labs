```python
import numpy as np
```

```python
def function (x_1:float, x_2:float, x_3:float):
  func = x_1 ** 2 + x_2 ** 2 + x_3 ** 2 + x_1 - 2 * x_2
  return func
```

```python
def derivative (x:float, y:float, z:float):
  df = np.array([2 * x + 1, 2 * y - 2, 2 * z])
  return df
```

```python
def new_point (x:float, y:float, z:float, precision:float):
  get_derivative = np.array([derivative(x, y, z)])
  next_point = np.array([x - (precision * get_derivative[0][0]), y - (precision * get_derivative[0][1]), z - (precision * get_derivative[0][2])])
  return next_point
```

```python
def check_accuracy (x:float, y:float, z:float):
  point = np.array([derivative(x, y, z)])
  x_check = point[0][0] ** 2
  y_check = point[0][1] ** 2
  z_check = point[0][2] ** 2
  return np.power(x_check + y_check + z_check, 0.5)
```

```python
precision = float(input("Введите желаемую точность: "))
x = float(input("Введите точку x: "))
y = float(input("Введите точку y: "))
z = float(input("Введите точку z: "))
```

```python
while True:
  if (check_accuracy(x, y, z) <= precision):
    break
  point = new_point(x, y, z, 0.001)
  x = point[0]
  y = point[1]
  z = point[2]
print(round(point[0], 3), " ", round(point[1], 3), " ", round(point[2], 3))
```
