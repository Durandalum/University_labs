```python
import random as rd
```

```python
def particles_generator(count:int):                              
    point = []
    speed = []
    for i in range(count):
        x_1 = rd.uniform(-10, 10)
        x_2 = rd.uniform(-10, 10)
        x_3 = rd.uniform(-10, 10)
        
        speed_1 = rd.uniform(-5, 5)
        speed_2 = rd.uniform(-5, 5)
        speed_3 = rd.uniform(-5, 5)
        
        point.append([x_1, x_2, x_3])
        speed.append([speed_1, speed_2, speed_3])
    return point, speed
```

```python
def particles_update(point:float, speed:float, pBest:float, gBest:float):      
    for i in range(len(point)):
        for j in range(len(speed[0])):
            speed[i][j] = 0.5 * speed[i][j] + 2.0 * rd.random() * (pBest[i][j] - point[i][j]) + 2.0 * rd.random() * (gBest[j] - point[i][j])
            point[i][j] += speed[i][j]
            if point[i][j] < -10:
                point[i][j] = -10
            if point[i][j] > 10:
                point[i][j] = 10
```

```python
def function (x_1:float, x_2:float, x_3:float):
  func = x_1 ** 2 + x_2 ** 2 + x_3 ** 2 + x_1 - 2 * x_2
  return func
```
