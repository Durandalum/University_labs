```python
import numpy as np
import random as r
```

```python
def mutation(point:float):
    if (r.random() > 0.5):
      point += 0.1
    else:
      point += -0.1
    return point
```

```python
def new_point (point_count:int):
  x_1 = r.random()
  x_2 = r.random()
  x_3 = r.random()
  function = x_1 ** 2 + x_2 ** 2 + x_3 ** 2 + x_1 - 2 * x_2
  point = Point(x_1, x_2, x_3, function)
  p = np.array([[point.x, point.y, point.z, point.func]])
  for i in range(1, point_count):
    x_1 = r.random()
    x_2 = r.random()
    x_3 = r.random()

    if (r.random() > 0.5):
      x_1 += 1
    if (r.random() > 0.5):
      x_2 += 1
    if (r.random() > 0.5):
      x_3 += 1

    if (r.random() < 0.5):
      x_1 *= -1
    if (r.random() < 0.5):
      x_2 *= -1
    if (r.random() < 0.5):
      x_3 *= -1
      
    function = x_1 ** 2 + x_2 ** 2 + x_3 ** 2 + x_1 - 2 * x_2
    point = Point(x_1, x_2, x_3, function)
    p = np.append(p, [[point.x, point.y, point.z, point.func]], axis = 0)
  return p
```

```python
class Point:
    def __init__(self, x, y, z, func):
        self.x = x
        self.y = y
        self.z = z
        self.func = func
```

```python
def byDF_key(point:float):
    return np.absolute(point[3])
```

```python
def iter(points_array:float, best_variants:float):
  best_variants = 0.4
  mutation_chance = 0.05
  x_1 = 0.0
  x_2 = 0.0
  x_3 = 0.0

  points_array = sorted(points_array, key = byDF_key)

  point_significancy = points_array[0][3] / (points_array[0][3] + points_array[1][3])

  if (r.random() <= point_significancy):
    x_1 = points_array[0][0]
  else:
    x_1 = points_array[1][0]

  if (r.random() <= point_significancy):
    x_2 = points_array[0][1]
  else:
    x_2 = points_array[1][1]

  if (r.random() <= point_significancy):
    x_3 = points_array[0][2]
  else:
    x_3 = points_array[1][2]

  if (r.random() <= mutation_chance):
    x_1 = mutation(x_1)
    x_2 = mutation(x_2)
    x_3 = mutation(x_3)

  function = x_1 ** 2 + x_2 ** 2 + x_3 ** 2 + x_1 - 2 * x_2
  new_points = Point(x_1, x_2, x_3, function)
  p = np.array([[new_points.x, new_points.y, new_points.z, new_points.func]])

  points_array = sorted(points_array, key = byDF_key)

  for i in range(1, int(len(points_array) * best_variants)):
    point_significancy = points_array[i][3] / (points_array[i][3] + points_array[i + 1][3])

    if (r.random() <= point_significancy):
      x_1 = points_array[i][0]
    else:
      x_1 = points_array[i + 1][0]

    if (r.random() <= point_significancy):
      x_2 = points_array[i][1]
    else:
      x_2 = points_array[i + 1][1]

    if (r.random() <= point_significancy):
      x_3 = points_array[i][2]
    else:
      x_3 = points_array[i + 1][2]

    if (r.random() <= mutation_chance):
      x_1 = mutation(x_1)
      x_2 = mutation(x_2)
      x_3 = mutation(x_3)

    function = x_1 ** 2 + x_2 ** 2 + x_3 ** 2 + x_1 - 2 * x_2
    new_points = Point(x_1, x_2, x_3, function)
    p = np.append(p, [[new_points.x, new_points.y, new_points.z, new_points.func]], axis = 0)
  
  random_child = 0
  clones_number = len(points_array) - len(p)
  for i in range(0, clones_number):
    random_child = r.choice(p)

    x_1 = random_child[0]
    x_2 = random_child[1]
    x_3 = random_child[2]

    x_1 = mutation(x_1)
    x_2 = mutation(x_2)
    x_3 = mutation(x_3)

    function = x_1 ** 2 + x_2 ** 2 + x_3 ** 2 + x_1 - 2 * x_2
    children_points = Point(x_1, x_2, x_3, function)
    p = np.append(p, [[children_points.x, children_points.y, children_points.z, children_points.func]], axis = 0)
  p = sorted(p, key = byDF_key)
  return p
```

```python
point_count = int(input("Введите желаемое количество точек: "))
precision = float(input("Введите желаемую точность: "))
best_value = float(input("Введите желаемый коэффициент лучших точек: "))
point = new_point(point_count)
points = np.array([[]])
points = iter(point, best_value)
max_iterations = 1000

err = 1
while (err > 1 - (precision / 100)):
  if (max_iterations == 0):
    break
  points = iter(point, best_value)
  error = 0

  for i in range(0, len(points)):
    error += np.absolute(points[i][3])
  error /= len(points)
  max_iterations -= 1
print()
print("Лучшая точка:")
print("X: ", round(points[0][0], 3))
print("Y: ", round(points[0][1], 3))
print("Z: ", round(points[0][2], 3))
print("Значение функции: ", points[0][3])
  
```
