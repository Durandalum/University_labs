```
datta = { 'country' : data['country'], 'points' : data['points'], 'price' : data['price'], 'province' : data['province']}
my_dataset = pd.DataFrame(datta, columns = ['country', 'points', 'price', 'province'])
```

```
my_dataset = my_dataset.dropna()
```

```
my_dataset.isnull().sum()
```

```
my_dataset.info()
```

```
my_dataset.describe()
```

```
my_dataset.plot(y = 'points', kind='hist', color='silver')
```

```
my_dataset.plot(y = 'price', kind='hist', color='indigo')
```

```
fig1, ax = plt.subplots()
ax.boxplot(my_dataset['price'], patch_artist=True, boxprops=dict(facecolor="tan"), showfliers=True)
boxlabel = mpatches.Patch(color = 'tan', label = 'price')
plt.legend(handles=[boxlabel])
```

```
fig1, bx = plt.subplots()
bx.boxplot(my_dataset['points'], patch_artist=True, boxprops=dict(facecolor="tan"), showfliers=False)
boxlabel = mpatches.Patch(color = 'tan', label = 'points')
plt.legend(handles=[boxlabel])
```

```
new_dataset = my_dataset.loc[my_dataset['price'] <= 50]
```

```
new_dataset.describe()
```

```
new_dataset.plot(y = 'price', kind='hist', color='indigo')
```

```
new_dataset.plot(y = 'points', kind='hist', color='silver')
```

```
fig1, ax = plt.subplots()
ax.boxplot(new_dataset['price'], patch_artist=True, boxprops=dict(facecolor="tan"), showfliers=True)
boxlabel = mpatches.Patch(color = 'tan', label = 'price')
plt.legend(handles=[boxlabel])
```

```
fig1, bx = plt.subplots()
bx.boxplot(new_dataset['points'], patch_artist=True, boxprops=dict(facecolor="tan"), showfliers=False)
boxlabel = mpatches.Patch(color = 'tan', label = 'points')
plt.legend(handles=[boxlabel])
```

```
plt.figure(figsize=(12,10), dpi= 100)
sns.heatmap(new_dataset.corr(), xticklabels=new_dataset.corr().columns, yticklabels=new_dataset.corr().columns, cmap='RdYlGn', center=0, annot=True)
plt.title('Корелляции признаков', fontsize=22)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```

```
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score
```

```
X = np.array(new_dataset[['price']])
y = np.array(new_dataset['points'])
```

```
regr = RandomForestRegressor(n_estimators = 500, max_features = 1)
cross_val_score(regr, X, y, cv=10).mean()
```
