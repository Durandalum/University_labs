```
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (15, 5)  
```

```
x = np.array([-1, 0, 1, 2, 3, 4, 5])
y = np.array([2.1, 2.2, 2.3, 2.4, 2.5, 2.5, 2.4])
```

```
factor_1, rss_1, *not_used = np.polyfit(x, y, 1, full='true')
factor_2, rss_2, *not_used = np.polyfit(x, y, 2, full='true')
factor_3, rss_3, *not_used = np.polyfit(x, y, 3, full='true')
factor_4, rss_4, *not_used = np.polyfit(x, y, 4, full='true')
factor_5, rss_5, *not_used = np.polyfit(x, y, 5, full='true')

#factor_1, factor_2, factor_3, factor_4, factor_5, factor_6, factor_7, factor_10 
rss_1, rss_2, rss_3, rss_4, rss_5
```

```
print("Cформируем полиномы:")
poly_1 = np.poly1d(factor_1)
print("Полином 1 порядка: \n", poly_1, '\n')
poly_2 = np.poly1d(factor_2)
print("Полином 2 порядка: \n", poly_2, '\n')
poly_3 = np.poly1d(factor_3)
print("Полином 3 порядка: \n", poly_3, '\n')
poly_4 = np.poly1d(factor_4)
print("Полином 4 порядка: \n", poly_4, '\n')
poly_5 = np.poly1d(factor_5)
print("Полином 5 порядка: \n", poly_5, '\n')
```

```
print('RSS полинома 1-го порядка:',poly_1(7))
print('RSS полинома 2-го порядка:',poly_2(7))
print('RSS полинома 3-го порядка:',poly_3(7))
print('RSS полинома 4-го порядка:',poly_4(7))
print('RSS полинома 5-го порядка:',poly_5(7))
```

```
x1 = np.array(range(-15, 15))
plt.plot(x, y, 'D', label='x')
plt.plot(x1, poly_1(x1), label='parabola_func-1')
plt.plot(x1, poly_2(x1), label='parabola_func-2')
plt.plot(x1, poly_3(x1), label='parabola_func-3')
plt.plot(x1, poly_4(x1), label='parabola_func-4')
plt.plot(x1, poly_5(x1), label='parabola_func-5')
plt.xlabel('x')
plt.ylabel('y')
plt.title("RSS")
plt.legend()
plt.show()
```
