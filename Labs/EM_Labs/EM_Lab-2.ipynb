```
df = data.copy()

for i in range(int(data.shape[0]*0.04)):
    item = rnd.randint(0, data.shape[0])
    df['Weight'][item] = 'NaN'
```

```
np.sum(np.isnan(df))
```

```
height = df['Height'].to_numpy()
weight = df['Weight'].to_numpy()
```

```
height_none=np.array(height[np.isnan(weight)])
```

```
height[~np.isnan(weight)]
```

```
df = df.dropna()
np.sum(np.isnan(df))
```

```
from scipy import interpolate
none_height_array = []
dataset = interpolate.interp1d(df['Height'],df['Weight']) 
none_height_array = dataset(height_none)
```

```
plt.plot(data['Height'], data['Weight'],'*')
plt.plot(height_none, none_height_array,'D')
```
