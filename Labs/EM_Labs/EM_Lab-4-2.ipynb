```
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (15, 5)  
```

```
x = np.array([-1, 0, 1, 2, 3, 4, 5])
y = np.array([2.1, 2.2, 2.3, 2.4, 2.5, 2.5, 2.4])
```

```
model_1 = np.polyfit(x, y, 1)
print(model_1)

model_2 = np.polyfit(x, y, 2)
print(model_2)

model_3 = np.polyfit(x, y, 3)
print(model_3)

model_4 = np.polyfit(x, y, 4)
print(model_4)

model_5 = np.polyfit(x, y, 5)
print(model_5)
```

```
print("Cформируем полином 1 порядка:")
poly_1 = np.poly1d(model_1)
print(poly_1,'\n')
print("Cформируем полином 2 порядка:")
poly_2 = np.poly1d(model_2)
print(poly_2,'\n')
print("Cформируем полином 3 порядка:")
poly_3 = np.poly1d(model_3)
print(poly_3,'\n')
print("Cформируем полином 4 порядка:")
poly_4 = np.poly1d(model_4)
print(poly_4,'\n')
print("Cформируем полином 5 порядка:")
poly_5 = np.poly1d(model_5)
print(poly_5,'\n')
```

```
Xpredict=6
print(poly_1(Xpredict))
```

```
RSS = 0
for i in range(len(x)):
  RSS += pow(model_1[0] * x[i] + model_1[1] - y[i] , 2)
print('RSS полинома 1-го порядка:',RSS)
```

```
X = np.linspace(-5,10,100)

fig, ax = plt.subplots()
ax.set_title('Полином 1 порядка')
ax.set_ylabel('Y')
ax.set_xlabel('X')
ax.plot(x, y, 'D', color = 'cyan')
ax.plot(X, poly_1(X), color = 'indigo')
```
