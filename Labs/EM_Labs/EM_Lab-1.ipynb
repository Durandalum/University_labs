```
import numpy as np
import scipy as sps
from scipy.interpolate import interp1d
import pandas as pd
import matplotlib.pyplot as plt
```

```
def interpolation (x, y, X):
    polynomial = 0
    for j in range(len(y)):
      numerator = 1
      denominator = 1
      for i in range(len(x)):
        if i == j:
          continue
        else:
          numerator *= (X - x[i])
          denominator *= (x[j] - x[i])
      polynomial += y[j] * numerator / denominator
    return polynomial
```

```
x = [50, 55, 60, 65, 70, 75]
y = [0.28, 0.31, 0.22, 0.04, -0.14, -0.27]
X = [52, 62, 73]
polynomial = []
for i in range(len(X)):
  polynomial.append((interpolation(x, y, X[i])))
print (polynomial)
```

```
X2 = [52, 62, 73]
polynom = []
for i in range(len(X)):
  f = interp1d(x, y)
  polynom.append(f(X[i]))
  print (polynom[i])
```
