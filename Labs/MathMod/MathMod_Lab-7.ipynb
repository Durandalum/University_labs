```
import numpy as np
import matplotlib.pyplot as plt
```

```
m_0 = float(0.55)
m_k = float(2.95)
dm = float(0.05)
p_0 = float(2.1)
p_k = float(6.9)
dp = float(0.1)
d_0 = float(2.4)
d_k = float(7.2)
dd = float(0.1)
m_2 = float(0.12)
N_0 = float(25000)
```

```
def frange(x, y, jump):
    while x < y:
        yield x
        x += jump
```

```
m = m_0
p = p_0
d = d_0

x = np.array([])
y = np.array([])
for N in range(24000, 72000, 1000):
  q = 1 / (m - m_2) * ((1 / (p * d)) * np.log(N_0 / N) - m_2)
  x = np.append(x, N)
  y = np.append(y, q)
  if (m < m_k):
    m += dm
  if (p < p_k):
    p += dp
  if (d < d_k):
    d += dd
```

```
fig, ax = plt.subplots()
ax.plot(x, y, label='line-1')
ax.grid()
ax.legend()
```
