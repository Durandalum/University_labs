```
import numpy as np
import matplotlib.pyplot as plt
```

```
m_0 = float(0.55)
m_k = float(2.95)
dm = float(0.05)
p_0 = float(2.1)
p_k = float(6.9)
dp = float(0.1)
d_0 = float(2.4)
d_k = float(7.2)
dd = float(0.1)
N_0 = float(24000)
d_N = float(72000)
dN = float(1000)
```

```
m = m_0
p = p_0
d = d_0
N = N_0

x_1 = np.array([])
y_1 = np.array([])
while m <= m_k:
  pulse_count = N * np.exp(-m * p * d)
  x_1 = np.append(x_1, m)
  y_1 = np.append(y_1, pulse_count)
  m += dm
  N += dN
```

```
m = m_0
p = p_0
d = d_0
N = N_0

x_2 = np.array([])
y_2 = np.array([])

while p <= p_k:
  pulse_count = N * np.exp(-m * p * d)
  x_2 = np.append(x_2, p)
  y_2 = np.append(y_2, pulse_count)
  p += dp
  N += dN
```

```
m = m_0
p = p_0
d = d_0
N = N_0

x_3 = np.array([])
y_3 = np.array([])

while d <= d_k:
  pulse_count = N * np.exp(-m * p * d)
  x_3 = np.append(x_3, d)
  y_3 = np.append(y_3, pulse_count)
  d += dd
  N += dN
```

```
fig, ax = plt.subplots()
ax.plot(x_1, y_1, label='line-1')
ax.grid()
ax.legend()
```

```
fig, bx = plt.subplots()
bx.plot(x_2, y_2, label='line-1')
bx.grid()
bx.legend()
```

```
fig, cx = plt.subplots()
cx.plot(x_3, y_3, label='line-1')
cx.grid()
cx.legend()
```

```
fig, dx = plt.subplots()
dx.plot(x_1, y_1, label='line-1')
dx.plot(x_2, y_2, label='line-2')
dx.plot(x_3, y_3, label='line-3')
dx.grid()
dx.legend()
```
