```
import numpy as np
import matplotlib.pyplot as plt
```

```
x_0 = float(18)
m_0 = float(200)
m_k = float(1100)
dm = float(450)
c_0 = float(20000)
c_k = float(40000)
dc = float(10000)
k_0 = float(3000)
k_k = float(3900)
dk = float(450)
x = float(3)
t_k = 5
dt = float(input("Введите шаг t:"))
```

```
def frange(x, y, jump):
    while x < y:
        yield x
        x += jump
```

```
k = k_0
c = c_0
m = m_0

t = 0
x_1 = np.array([])
y_1 = np.array([])
x_1 = np.append(x_1, t)
y_1 = np.append(y_1, x_0)
for i in frange (t + dt, t_k, dt):
  x_i = i * dt + x_0
  x_i = (-k * x_i - c * x) / m * dt + x_i
  x_1 = np.append(x_1, i)
  y_1 = np.append(y_1, x_i)
k += dk
c += dc
m += dm
```

```
t = 0
x_2 = np.array([])
y_2 = np.array([])
x_2 = np.append(x_2, t)
y_2 = np.append(y_2, x_0)
for i in frange (t + dt, t_k, dt):
  x_i = i * dt + x_0
  x_i = (-k * x_i - c * x) / m * dt + x_i
  x_2 = np.append(x_2, i)
  y_2 = np.append(y_2, x_i)
k += dk
c += dc
m += dm
```

```
t = 0
x_3 = np.array([])
y_3 = np.array([])
x_3 = np.append(x_3, t)
y_3 = np.append(y_3, x_0)
for i in frange (t + dt, t_k, dt):
  x_i = i * dt + x_0
  x_i = (-k * x_i - c * x) / m * dt + x_i
  x_3 = np.append(x_3, i)
  y_3 = np.append(y_3, x_i)
k += dk
c += dc
m += dm
```

```
plt.plot(x_1, y_1, label='line-1')
plt.plot(x_2, y_2, label='line-2')
plt.plot(x_3, y_3, label='line-3')
plt.legend()
plt.show()
```
