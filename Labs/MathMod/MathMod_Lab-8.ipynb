```
import numpy as np
import matplotlib.pyplot as plt
import math as m
```

```
d_0 = float(4)
d_k = float(12)
dd = float(0.2)
a_0 = float(20)
a_k = float(190)
da = float(5)
R_0 = float(6.5)
R_k = float(26.5)
dR = float(0.5)
h_0 = float(2.25)
h_k = float(12.25)
dh = float(0.25)
Q = float(25)
```

```
def frange(x, y, jump):
    while x < y:
        yield x
        x += jump
```

```
d = d_0
a = a_0
R = R_0
h = h_0


x = np.array([])
y = np.array([])
for h in frange(h, h_k, dh):
  B = (Q * m.pow(d, 2)) / 64 * m.tan((d * a) / 2)
  N = B * (h * m.pow(m.cos(a), 3)) / (m.pow(R, 2) * pow(m.cos(a), 2) - R * h * m.pow(m.sin(a), 2) + m.pow(h, 2)) ** 1.5
  x = np.append(x, h) 
  y = np.append(y, N)
  d += dd
  a += da
  R += dR
```

```
fig, ax = plt.subplots()
ax.plot(x, y, label='line-1')
ax.grid()
ax.legend()
```
