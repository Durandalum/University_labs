```
import numpy as np
import matplotlib.pyplot as plt
```

```
q = float(2500)
l_0 = float(200)
l_k = float(1100)
dl = float(450)
r_0 = float(3400)
r_k = float(4000)
dr = float(300)
c_0 = float(2800)
c_k = float(3800)
dc = float(500)
q_0 = 3
E = float(2000)
t_k = 5
dt = float(input("Введите шаг t:"))
```

```
def frange(x, y, jump):
    while x < y:
        yield x
        x += jump
```

```
l = l_0
r = r_0
c = c_0

t = 0
Q = q
x_1 = np.array([])
y_1 = np.array([])
x_1 = np.append(x_1, t)
y_1 = np.append(y_1, Q)
for i in frange (t + dt, t_k, dt):
  Q = Q * dt * q_0
  Q = (E - (1 / c) * Q - (r * Q)) / l * dt + Q
  x_1 = np.append(x_1, i)
  y_1 = np.append(y_1, Q)
l += dl
r += dr
c += dc
```

```
t = 0
Q = q
x_2 = np.array([])
y_2 = np.array([])
x_2 = np.append(x_2, t)
y_2 = np.append(y_2, Q)
for i in frange (t + dt, t_k, dt):
  Q = Q * dt * q_0
  Q = (E - (1 / c) * Q - (r * Q)) / l * dt + Q
  x_2 = np.append(x_2, i)
  y_2 = np.append(y_2, Q)
l += dl
r += dr
c += dc
```

```
t = 0
Q = q
x_3 = np.array([])
y_3 = np.array([])
x_3 = np.append(x_3, t)
y_3 = np.append(y_3, Q)
for i in frange (t + dt, t_k, dt):
  Q = Q * dt * q_0
  Q = (E - (1 / c) * Q - (r * Q)) / l * dt + Q
  x_3 = np.append(x_3, i)
  y_3 = np.append(y_3, Q)
```

```
fig, ax = plt.subplots()
ax.plot(x_1, y_1, label='line-1')
ax.legend()
```
