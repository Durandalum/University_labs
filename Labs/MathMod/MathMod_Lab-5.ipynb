```
import numpy as np
import matplotlib.pyplot as plt
```

```
x1_0 = float(2)
y1_0 = float(2)
x_0 = float(2)
y_0 = float(2)
m_0 = float(1000)
m_k = float(1700)
dm = float(350)
c_0 = float(200000)
c_k = float(250000)
dc = float(25000)
k_0 = float(10000)
k_k = float(10700)
dk = float(350)
dt = float(input("Введите шаг t:"))
t_k = 2
g = 9.8
```

```
def frange(x, y, jump):
    while x < y:
        yield x
        x += jump
```

```
m = m_0
k = k_0
c = c_0

x_speed_1 = np.array([])
x_position_1 = np.array([])
y_speed_1 = np.array([])
y_position_1 = np.array([])

for i in frange (0, t_k, dt):
  x_new = x_0 + dt * x1_0
  y_new = y_0 + dt * y1_0

  f = c * (y_0 - x_0)
  x1_new = x1_0 - (g + k * (x1_0 / m) - (f/m)) * dt
  y1_new = y1_0 - (g + (f/m)) * dt

  x1_0 = x1_new
  y1_0 = y1_new
  x_0 = x_new
  y_0 = y_new
  x_speed_1 = np.append(x_speed_1, i)
  x_position_1 = np.append(x_position_1, x1_0)
  y_speed_1 = np.append(y_speed_1, i)
  y_position_1 = np.append(y_position_1, y1_0)
```

```
m += dm
c += dc
k += dk

x1_0 = float(2)
y1_0 = float(2)
x_0 = float(2)
y_0 = float(2)

x_speed_2 = np.array([])
x_position_2 = np.array([])
y_speed_2 = np.array([])
y_position_2 = np.array([])


for i in frange (0, t_k, dt):
  x_new = x_0 + dt * x1_0
  y_new = y_0 + dt * y1_0

  f = c * (y_0 - x_0)
  x1_new = x1_0 - (g + k * (x1_0 / m) - (f/m)) * dt
  y1_new = y1_0 - (g + (f/m)) * dt

  x1_0 = x1_new
  y1_0 = y1_new
  x_0 = x_new
  y_0 = y_new
  x_speed_2 = np.append(x_speed_2, i)
  x_position_2 = np.append(x_position_2, x1_0)
  y_speed_2 = np.append(y_speed_2, i)
  y_position_2 = np.append(y_position_2, y1_0)
```

```
m += dm
c += dc
k += dk

x1_0 = float(2)
y1_0 = float(2)
x_0 = float(2)
y_0 = float(2)

x_speed_3 = np.array([])
x_position_3 = np.array([])
y_speed_3 = np.array([])
y_position_3 = np.array([])


for i in frange (0, t_k, dt):
  x_new = x_0 + dt * x1_0
  y_new = y_0 + dt * y1_0

  f = c * (y_0 - x_0)
  x1_new = x1_0 - (g + k * (x1_0 / m) - (f/m)) * dt
  y1_new = y1_0 - (g + (f/m)) * dt

  x1_0 = x1_new
  y1_0 = y1_new
  x_0 = x_new
  y_0 = y_new
  x_speed_3 = np.append(x_speed_3, i)
  x_position_3 = np.append(x_position_3, x1_0)
  y_speed_3 = np.append(y_speed_3, i)
  y_position_3 = np.append(y_position_3, y1_0)
```

```
fig, ax = plt.subplots()
ax.plot(x_speed_1, x_position_1, label='line-1')
ax.plot(x_speed_2, x_position_2, label='line-2')
ax.plot(x_speed_3, x_position_3, label='line-3')
ax.legend()
```

```
fig, bx = plt.subplots()
bx.plot(y_speed_1, y_position_1, label='line-1')
bx.plot(y_speed_2, y_position_2, label='line-2')
bx.plot(y_speed_3, y_position_3, label='line-3')
bx.legend()
```
