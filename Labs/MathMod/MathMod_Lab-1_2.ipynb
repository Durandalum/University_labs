```
import numpy as np
import matplotlib.pyplot as plt
```

```
a = float(0.4)
c_1 = float(1.8)
c_2 = float(3.6)
t_0 = float(0)
t_k = float(1.1)
dt = float(0.1)
f_0 = float(5)
f_k = float(10)
df = float(2)
f = f_0
```

```
def frange(x, y, jump):
    while x < y:
        yield x
        x += jump
```

```
t = np.array([])
x_t = np.array([])
for i in frange (t_0, t_k, dt):
  omega = 2 * np.pi * f
  x = np.exp(-a * i) * (c_1 * np.cos(omega * i) + c_2 * np.sin(omega * i))
  t = np.append(t, i)
  x_t = np.append(x_t, x)
  if (f <= f_k):
    f += df
```

```
plt.plot(t, x_t, label='function')
```
