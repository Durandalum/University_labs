```python
import numpy as np
```

```python
Fe1 = np.array([65.7, 66.1, 63.8, 64.3, 64.7, 64.9, 65.5, 66.7, 63.5, 65.1, 61.7, 62.8])
Fe2 = np.array([65.38, 65.95, 63.93, 64.61, 65.1, 65.06, 65.6, 66.99, 63.41, 65.05, 61.96, 62.47])
fault_rel = 0
fault_abs = 0
```

```python
Fe1_aver = Fe1.sum() / len(Fe1)
Fe2_aver = Fe2.sum() / len(Fe2)
Fe_total_aver = (Fe1_aver + Fe2_aver) / 2
print("Среднее значение по 2 выборкам: ", round(Fe_total_aver, 3))
```

```python
sum_fault = 0
for i in range(0, len(Fe1)):
  sum_fault += np.absolute(Fe1[i] - Fe2[i])
fault_abs = sum_fault / len(Fe1)
print("Абсолютная погрешность:", round(fault_abs, 3))
```

```python
fault_rel = fault_abs / Fe2_aver * 100
print("Относительная погрешность:",round(fault_rel, 3), "%")
```
