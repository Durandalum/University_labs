```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = np.array([5000, 10000, 15000])
fig, ax = plt.subplots()
for i in range (0, 3):
  x = np.array([])
  y = np.array([])
  plot_x = np.array([])
  plot_y = np.array([])
  for t in range(1, 61):
    fault = 100/np.sqrt(N[i] * t)
    x = np.append(x, t)
    y = np.append(y, fault)
  plot_x = np.append(plot_x, x)
  plot_y = np.append(plot_y, y)
  ax.plot(plot_x, plot_y, label='line-1')
  ax.grid()
  ax.legend()
```
