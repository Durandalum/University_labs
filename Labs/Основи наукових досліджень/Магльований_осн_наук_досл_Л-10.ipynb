```python
import numpy as np
```

```python
x = np.array([65.4, 60.9, 66.64, 64.6, 64.8, 65.5, 64.6, 63.76, 62.0, 61.67])
y = np.array([67.46, 60.76, 65.66, 64.79, 64.06, 66.96, 65.4, 63.09, 62.64, 61.66])
```

```python
def dispersion (array):
  result = 0
  average = np.sum(array) / len(array)
  for i in range(0, len(array)):
    result += np.power(array[i] - average, 2)
  result = result / len(array)
  return result
```

```python
x_dispersion = dispersion(x)
y_dispersion = dispersion(y)

f_test = 0

if (x_dispersion > y_dispersion): 
  f_test = x_dispersion / y_dispersion
else:
  f_test = y_dispersion / x_dispersion

print("Критерий Фишера: ", round(f_test, 3))
```
