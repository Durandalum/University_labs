```python
import numpy as np
```

```python
x = np.array([29.41, 31.94, 30.14, 22.6,  24.8,  30.87, 24.2,  28.73, 28.11, 31.27, 30.4, 24.11, 21.27, 28.45, 27.61, 31.07, 21.27, 28.66, 30.54, 26.88])
y = np.array([29.41, 30.02, 27.32, 23.79, 24.02, 28.36, 25.48, 28.09, 26.64, 33.33, 29.9, 25.77, 22.35, 27.43, 27.0, 31.02, 22.35, 25.15, 29.97, 26.6 ])
```

```python
def dispersion (array):
  result = 0
  average = np.sum(array) / len(array)
  for i in range(0, len(array)):
    result += np.power(array[i] - average, 2)
  result = result / len(array)
  return result
```

```python
def correlation (x_array, y_array, x_deviation, y_deviation):
  multiplication = np.sum(x_array * y_array) / len(x_array)
  x_average = np.sum(x_array) / len(x_array)
  y_average = np.sum(y_array) / len(y_array)
  result = (multiplication - (x_average * y_average)) / (x_deviation * y_deviation)
  return result
```

```python
def confidence_interval (x_array, y_array, x_deviation, y_deviation):
  x_average = np.sum(x_array) / len(x_array)
  y_average = np.sum(y_array) / len(y_array)
  x_confidence_from = round((x_average - 1.96 * (x_deviation / np.sqrt(len(x_array)))), 3)
  x_confidence_to = round((x_average + 1.96 * (x_deviation / np.sqrt(len(x_array)))), 3)
  y_confidence_from = round((y_average - 1.96 * (y_deviation / np.sqrt(len(y_array)))), 3)
  y_confidence_to = round((y_average + 1.96 * (y_deviation / np.sqrt(len(y_array)))), 3)
  print("Доверительный интервал X:")
  print("От: ", x_confidence_from)
  print("До: ", x_confidence_to)
  print(" ")
  print("Доверительный интервал Y: ")
  print("От: ", y_confidence_from)
  print("До: ", y_confidence_to)
```

```python
x_dispersion = dispersion(x)
y_dispersion = dispersion(y)
print("Дисперсия X: ", round(x_dispersion, 3))
print("Дисперсия Y: ", round(y_dispersion, 3))

x_deviation = np.sqrt(x_dispersion)
y_deviation = np.sqrt(y_dispersion)
print("Среднеквадратичное отклонение X: ", round(x_deviation, 3))
print("Среднеквадратичное отклонение Y: ", round(y_deviation, 3))

correlation_coefficient = correlation(x, y, x_deviation, y_deviation)
print("Коэффициент корреляции: ", round(correlation_coefficient, 3))

interval = confidence_interval(x, y, x_deviation, y_deviation)
```
