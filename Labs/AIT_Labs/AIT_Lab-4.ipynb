```
import numpy as np
import pandas as pd
from __future__ import print_function
from genetic_selection import GeneticSelectionCV
from sklearn import linear_model
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
```

```
data = pd.read_csv('/content/drive/MyDrive/Advertising.csv')
```

```
x = np.array(data[['TV', 'Radio', 'Newspaper']])
y = np.array(data['Sales'])
```

*Тренировка на учебных данных.*

```
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.7, random_state=0)
estimator = linear_model.LinearRegression()
selector = GeneticSelectionCV(estimator, 2, crossover_proba= 0.5)
selector = selector.fit(x_train, y_train)
train_predict = selector.predict(x_train)
metrics.mean_squared_error(y_train, train_predict)
```

*Валидация на тестовых данных.*

```
test_predict = selector.predict(x_test)
metrics.mean_squared_error(y_test, test_predict)
```

```
scores = cross_val_score(selector, x_test, y_test)
print(scores)
```
