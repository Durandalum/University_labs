```
import numpy as np
import scipy as sp
from scipy.optimize import brentq
import matplotlib.pyplot as plt
%matplotlib inline
```

Уравнение №1, cos(x + 0,3) = x  ^ 2.

```
x = np.linspace(-3, 3, 100)

answer = np.cos(x + 0.3) - x ** 2
answer
```

```
def f(x):
    return np.cos(x + 0.3) - x ** 2

plt.plot(x, f(x))
plt.axhline(color = 'black')
root, info = brentq(f, -2, 0, full_output=True)
root, info.converged
```

Уравнение №2, 2 * x ^ 4 + 8 * x ^ 3 + 8 * x ^ 2 - 1.

```
x = sp.linspace(-5, 5, 30)

answer = 2 * x ** 4 + 8 * x ** 3 + 8 * x ** 2 - 1
answer
```

```
def f(x):
    return 2 * x ** 4 + 8 * x ** 3 + 8 * x ** 2 - 1

plt.plot(x, f(x))
plt.axhline(color = 'k')
root, info = brentq(f, -4, 0, full_output=True)
root, info.converged
```

Уравнение №3, 2 * sin(x + pi / 3) = 0,5 * x ^ 2 - 1.

```
x = sp.linspace(0, 3, 50)

answer = 2 * np.sin(x + np.pi / 3) - 0.5 * x ** 2 + 1
answer
```

```
def f(x):
    return 2 * np.sin(x + np.pi / 3) - 0.5 * x ** 2 + 1

plt.plot(x, f(x))
plt.axhline(color = 'k')
root, info = brentq(f, 1, 2, full_output=True)
root, info.converged
```

Уравнение №4, -2 * lg(x) - x / 2 + 1.

```
x = sp.linspace(1, 5, 65)

answer = -2 * np.log10(x) - x / 2 + 1
answer
```

```
def f(x):
    return -2 * np.log10(x) - x / 2 + 1

plt.plot(x, f(x))
plt.axhline(color = 'k')
root, info = brentq(f, 1, 2, full_output=True)
root, info.converged
```
