```
# описание функции
def du (x,y):
    return x+cos(y/sqrt(10))

def fx (x_0,y_0,x_k,n_1):
# шаг интегрирования
    h= (x_k-x_0) /n_1
    print('x[', 0, '] = ', round(x0,3), '|','y[', 0, '] = ', round(y0,3))
#Метод Эйлера
    for i in range (0,n_1):
        y1=y_0+h*du (x_0,y_0)
        x1=x_0+h
        print('x[', i+1, '] = ', round(x1,3), '|','y[', i+1, '] = ', round(y1,3))
        x_0=x1
        y_0=y1
    return y1

# начальные условия
print("Введите границы отрезка [x0; xk]")
x0 = float(input("x0 = "))
xk = float(input("xk = "))
print("Введите начальное условие y(0)")
y0 = float(input("y0 = "))
print("Введите количество разбиений отрезка:")
n = int(input("n = "))

print("Результат:")
y1 = "%.3f" % fx (x0,y0,xk,n)
print(y1)
```
