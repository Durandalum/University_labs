```
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (15, 5)
from random import *
```

```
N = int(input("Введите количество точек в полиноме: "))
```

```
x = [randint(-5, 5) for x in range(N)]
print(x)
y = [randint(-2, 3) for x in range(N)]
print(y)
m = 3
```

```
model = np.polyfit(x, y, m)
print(model)
```

```
print("Cформируем полином:")
poly = np.poly1d(model)
print(poly,'\n')
```

```
RSS = 0
for i in range(len(x)):
  RSS += pow(model[0] * x[i] + model[1] - y[i] , 2)
print('RSS полинома:', RSS)
```

```
X = np.arange(x[0] - 5, len(x), 0.5)

fig, ax = plt.subplots()
ax.set_ylabel('Y')
ax.set_xlabel('X')
ax.plot(x, y, 'o', color = 'cyan')
ax.plot(X, poly(X), color = 'indigo')
```
