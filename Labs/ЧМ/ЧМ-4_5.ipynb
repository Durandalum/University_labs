```
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```
def func(x):
    #return 2 * np.sin(x + np.pi / 3) - 0.5 * x ** 2 + 1
    #return x ** 2 - 3 * x + 2
    return 2 * x ** 4 + 8 * x ** 3 + 8 * x ** 2 - 1
```

```
def find_M(a, b, fun):
    h = 10**(-5)
    eps = 10**(-2)
    f_all = []
    for i in range(a, b + 1, 1):
        f_all.append((func(i + h) - func(i - h))/(2 * h))
    M = abs(max(f_all))
    return M
```

```
def iterations(a, b, M, eps, max_iter, fun):
    it = 0
    h = 10**-5
    x = a
    while it < max_iter:
        y1 = (fun(x + h) - fun(x)) / h
        it = it + 1
        if abs(fun(x)) < eps:
            return x
        else:
            if y1 > 0:
                x = x - fun(x) / M
            else:
                x = x + fun(x) / M
    print('Слишком много итераций')
```

```
# число M такое, что при этом выполняется неравенство 
# M > 0 и M > |f'(x)|
# для любого х из отрезка [a,b]
M = find_M(-4, 3, func)
print('M =', M)
res = iterations(1, 2, M, 10**-8, 10000, func)
print('Корень =', res)
x = np.linspace(-3, 2, 100)
plt.plot(x, func(x))
plt.axhline(color = 'k')
```
