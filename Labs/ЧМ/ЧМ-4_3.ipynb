```
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```
def func(x):
    #return 2 * np.sin(x + np.pi / 3) - 0.5 * x ** 2 + 1
    #return x ** 2 - 3 * x + 2
    return np.cos(x + 0.3) - x ** 2
```

```
def find_M(a, b, fun):
    h = 10**(-5)
    eps = 10**(-2)
    f_all = []
    for i in range(a, b + 1, 1):
        f_all.append((func(i + h) - func(i - h))/(2 * h))
    M = abs(max(f_all))
    return M
```

```
def rybakov(a, b, M, eps, funс):
    x = a
    h = 10**-5
    roots = []
    while x < b:
        # производная от x
        y1 = (funс(x + h) - funс(x - h))/(2 * h)
        if abs(y1) > M:
            print('|y1| > M')
            return
        else:
            if abs(funс(x)) <= eps:
                roots.append(x)
                x = x + eps
            else:
                x = x + (abs(funс(x)) / M)
    return roots
```

```
# число M такое, что при этом выполняется неравенство 
# M > 0 и M > |f'(x)|
# для любого х из отрезка [a,b]
M = find_M(-4, 3, func)
print('M =', M)
res = rybakov(-3, 3, 100, 10**-8, func)
print('Корни:', res)
x = np.linspace(-3, 3, 100)
plt.plot(x, func(x))
plt.axhline(color = 'k')
```
