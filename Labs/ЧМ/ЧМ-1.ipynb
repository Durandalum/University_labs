```
def cube_equatation (x3_ratio, x2_ratio, x_ratio):
  q = float((pow(x3_ratio, 2) - 3 * x2_ratio) / 9)
  r = float((2 * pow(x3_ratio, 3) - 9 * x3_ratio * x2_ratio) + 27 * x_ratio / 54)
  
  x1 = 0.0
  x2 = 0.0
  x3 = 0.0
  
  if (pow(r, 2) < pow(q, 3)):
    t = float(acos(r / sqrt(pow(q, 3))) / 3)
    x1 = -2 * sqrt(q) * cos(t) - x3_ratio / 3
    x2 = -2 * sqrt(q) * cos(t + (2 * pi / 3)) - x3_ratio / 3
    x3 = -2 * sqrt(q) * cos(t - (2 * pi / 3)) - x3_ratio / 3
    print("X =", x1)
    print("X =", x2)
    print("X =", x3)
  else:
    A = float(-pow(fabs(r) + sqrt(pow(r, 2) - pow(q, 3)), (1.0 / 3.0)))
    if (A != 0):
      B = float(q / A)
    else:
      B = 0.0
    
    x1 = (A + B) - x3_ratio / 3
    x2 = -(A + B) / 2 - (x3_ratio / 3)
    x3 = -(A + B) / 2 - (x3_ratio / 3)
    
    if (A == B):
      x2 = -A - x3_ratio / 3
      print("X =", x1)
      print("X =", x2)
    else:
      print("X =", x1)
      print("X =", x2, "+i")
      print("X =", x3, "-i")
```

```
def mini_cube_equatation(x3_ratio, num):
    print("Уравнение: ", x3_ratio, "X^3 + ", num)
    num /= x3_ratio
    x = float(pow(num, float(1)/3))
    print("X =", round(x))
```
