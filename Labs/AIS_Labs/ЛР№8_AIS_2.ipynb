```python
import numpy as np
import pandas as pd
import re
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import GaussianNB
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
def text_editor(text):
  text = re.sub(r'\d+', '', text)
  text = re.sub(r'[^\w\s]','', text)
  text = text.lower() 
  return text
```

```python
spam_ham_file = open('/content/drive/My Drive/SMSSpamCollection.txt', 'r')
spam_ham_categories = []
line_arr = []
for line in spam_ham_file:
  line_text = line.strip().split('\t')
  if line_text[0] == 'ham':
    spam_ham_categories.append(0)
  else:
    spam_ham_categories.append(1)
  line_arr.append(text_editor(line_text[1]))
```

```python
for i in range(10):
  print(line_arr[i])
```

Создайте мешок слов

```python
#Ваш код
tokenization = CountVectorizer()
tokenization.fit(line_arr)

bag_of_words = tokenization.transform(line_arr).toarray()
frequency_matrix = pd.DataFrame(bag_of_words, columns = tokenization.get_feature_names())
```

Реализуйте наивный Байесовский классификатор

```python
#Ваш код
tokenization = CountVectorizer()
data = tokenization.fit_transform(line_arr)

model = GaussianNB()
model.fit(data.toarray(), spam_ham_categories)
```

Определите принадлежность к классу произвольного смс-сообщения, которое не участвовало в обучении


```python
#Ваш код
is_it_ham = tokenization.transform([text_editor("I think electric cars are a great invention. It would be nice to buy one.")])
is_it_spam = tokenization.transform([text_editor("You have still not claimed the compensation you are due for the accident you had. To start the process please reply YES. To opt out text STOP")])

print(model.predict_proba(is_it_ham.toarray()))
print(model.predict_proba(is_it_spam.toarray()))

print(model.predict(is_it_ham.toarray()))
print(model.predict(is_it_spam.toarray()))
```

```python
import numpy as np
sentens = ['предоставляю','услуги','бухгалтера','спешите','купить','велосипед','надо','купить','молоко']
answer = ['spam','spam','ham']

from sklearn import preprocessing

le = preprocessing.LabelEncoder()
x=np.array(le.fit_transform(sentens)).reshape((3,3))
label=np.array(le.fit_transform(answer)).reshape((3,))

print (x)
print (label)

```

```python
from sklearn.naive_bayes import GaussianNB

model = GaussianNB()

model.fit(x,label)

predicted= model.predict([[4,2,9]]) #надо(4), купить(2), диван(9 - отсутствует в коллекции слов)
print ("Predicted Value:", predicted)
```
