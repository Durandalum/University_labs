```python
file = open('/content/drive/My Drive/Colab_Notebooks/sentences.txt', 'r')
sentences = file.read().lower()
words = nltk.sent_tokenize(sentences)
words_ = [nltk.word_tokenize(i) for i in words]
#words_
```

```python
words_ = [[j for j in i if (j not in string.punctuation)] for i in words_]
stop_words = stopwords.words('english')
words_ = [[j for j in i if (j not in stop_words)] for i in words_]
words_ = [[PorterStemmer().stem(j) for j in i] for i in words_]
words_ = [[j for j in i if j.isalpha()] for i in words_]
#words_[0][1]
#words__ = [set(words_[i]) for i in range(0, len(words_))]
#words__
```

```python
dic = {}
for i in words_:
    for j in i:
        if j not in dic:
            dic[j] = 1
        else:
            dic[j] +=1
dic = [i for i in dic if (dic[i] > 1)]
#words_[5], words_[6]
```

```python
matrix = np.zeros((len(sentences.split('\n')), len(dic)))
```

```python
for i in range(len(words_)):
    #print(i)
    for j in range(len(dic)):
        #print(j)
        #print(sentence[i].count(a[j]))
        if words_[i].count(dic[j]) == 0:
            matrix[i][j] = 0
        else:
            matrix[i][j] = words_[i].count(dic[j])
pd.DataFrame(data = matrix, columns = dic) 
```
