```python
import warnings
warnings.filterwarnings('ignore')
import numpy as np
%pylab inline
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

###**Генерация данных**

```python
plt.figure(figsize=(8, 8))
X, Y = make_blobs(n_features=2, centers=3, cluster_std = 2, random_state=42)
plt.scatter(X[:, 0], X[:, 1])
```

#**K - means**

```python
from sklearn.cluster import KMeans
y_pred = KMeans(n_clusters=3, random_state=42).fit_predict(X)
plt.figure(figsize=(8, 8))
plt.scatter(X[:, 0], X[:, 1], c=y_pred)
```

#**AgglomerativeClustering**

```python
from sklearn.cluster.hierarchical import AgglomerativeClustering

y_pred = AgglomerativeClustering(n_clusters=3, affinity='cosine', linkage='complete').fit_predict(X)
plt.figure(figsize=(8, 8))
plt.scatter(X[:, 0], X[:, 1], c=y_pred)
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage

Z = linkage(X, 'ward')
fig = plt.figure(figsize=(25, 10))
dn = dendrogram(Z)
```

```python
import numpy as np
import math
import operator
import matplotlib.pyplot as plt
```

```python
bloodArr = np.array([[0, 0], [0, 1], [0, 2], [0, 3], [1, 0], [1, 1], [1, 2], [1, 3], [2, 0], [2, 1], [2, 2], [2, 3], [3, 0], [3, 1], [3, 2], [3, 3]])
```

```python
plt.scatter(bloodArr[:, 0], bloodArr[:, 1])
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage

hierarchy = linkage(bloodArr, 'ward')
dndrgrm = dendrogram(hierarchy, orientation='bottom')
```

```python
from sklearn import mixture
#EM-алгоритм
y_pred = mixture.GaussianMixture(n_components = 2).fit_predict(bloodArr)
plt.scatter(bloodArr[:, 0], bloodArr[:, 1], c = y_pred);
```

```python
from sklearn.cluster import MiniBatchKMeans
y_pred = MiniBatchKMeans(n_clusters = 2, random_state = 4).fit_predict(bloodArr)
plt.scatter(bloodArr[:, 0], bloodArr[:, 1], c = y_pred)
```

```python
from sklearn.cluster.hierarchical import AgglomerativeClustering

y_pred = AgglomerativeClustering(n_clusters = 2, linkage = 'ward').fit_predict(bloodArr)
plt.scatter(bloodArr[:, 0], bloodArr[:, 1], c = y_pred)
```

```python
from sklearn.cluster import Birch
y_pred = Birch(n_clusters = 2).fit_predict(bloodArr)
plt.scatter(bloodArr[:, 0], bloodArr[:, 1], c = y_pred)
```

```python
from sklearn.cluster import SpectralClustering
y_pred = SpectralClustering(n_clusters = 2).fit_predict(bloodArr)
plt.scatter(bloodArr[:, 0], bloodArr[:, 1], c = y_pred)
```
