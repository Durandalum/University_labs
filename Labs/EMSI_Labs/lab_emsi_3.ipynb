```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (15, 5)  
```

```python
x = np.array([-1, 0, 1, 2, 3, 4, 5])
y = np.array([-6.1, -5.8, -5.2, -4.8, -4.5, -5.0, -5.2])
x_predict = 7
```

```python
func_1 = np.array([[len(x), x.sum()], [x.sum(), (x**2).sum()]])
func_y_1 = np.array([y.sum(), (x*y).sum()])
a0_1, a1_1 = np.linalg.solve(func_1, func_y_1)
a0_1, a1_1
```

```python
func_2 = np.array([[len(x), x.sum(), (x**2).sum()], [x.sum(), (x**2).sum(), (x**3).sum()], [(x**2).sum(), (x**3).sum(), (x**4).sum()]])
func_y_2 = np.array([y.sum(), (x*y).sum(), (((x**2)*y).sum())])
a0_2, a1_2, a2_2 = np.linalg.solve(func_2, func_y_2)
a0_2, a1_2, a2_2
```

```python
y_predict_1 = a0_1 + a1_1*x_predict
y_predict_2 = a0_2 + a1_2*x_predict + a2_2*(x_predict**2)
y_predict_1, y_predict_2
```

```python
rss1 = ((a0_1 + a1_1*x - y)**2).sum()
rss2 = ((a0_2 + a1_2*x + a2_2*(x**2) - y)**2).sum()
rss1, rss2
```

```python
x1 = np.array(range(-10, 10))
plt.plot(x, y, 'o', label='x')
plt.plot(x1, (a0_1 + a1_1*x1), label='linear')
plt.plot(x1, (a0_2 + a1_2*x1 + a2_2*(x1**2)), label='parabola')
plt.legend()
plt.show()
```
