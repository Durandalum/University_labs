```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([29.41, 31.94, 30.14, 22.6,  24.8,  30.87, 24.2,  28.73, 28.11, 31.27, 30.4, 24.11, 21.27, 28.45, 27.61, 31.07, 21.27, 28.66, 30.54, 26.88])
```

```python
def dispersion (array):
  result = 0
  average = np.sum(array) / len(array)
  for i in range(0, len(array)):
    result += np.power(array[i] - average, 2)
  result = result / len(array)
  return result
```

```python
def confidence_interval (x_array, x_deviation):
  x_average = np.sum(x_array) / len(x_array)
  x_confidence_from = round((x_average - 1.96 * (x_deviation / np.sqrt(len(x_array)))), 3)
  x_confidence_to = round((x_average + 1.96 * (x_deviation / np.sqrt(len(x_array)))), 3)
  return x_confidence_from, x_confidence_to
```

```python
x_dispersion = dispersion(x)
print("Дисперсия: ", round(x_dispersion, 3))

x_deviation = np.sqrt(x_dispersion)
print("Среднеквадратичное отклонение: ", round(x_deviation, 3))
interval_from, interval_to = confidence_interval(x, x_deviation)
  
print("Доверительный интервал:")
print("От: ", interval_from)
print("До: ", interval_to)

plot_x = np.array([])
plot_y = np.array([])
plot_interval_from = np.array([])
plot_interval_to = np.array([])
fig, ax = plt.subplots()
for i in range(0, len(x)):
  plot_x = np.append(plot_x, i)
  plot_y = np.append(plot_y, x[i])
  plot_interval_from = np.append(plot_interval_from, interval_from)
  plot_interval_to = np.append(plot_interval_to, interval_to)
ax.plot(plot_x, plot_y, color = 'green')
ax.plot(plot_x, plot_interval_from, color = 'purple')
ax.plot(plot_x, plot_interval_to, color = 'red')
ax.plot()
```
