```
def newton_simpson (a:float , b:float, count:float):
  width = float((b - a) / count)
  x_1 = float(0)
  x_2 = float(0)
  sum_1 = float(0)
  sum_2 = float(0)
  integral_simpson = float(0)

  for i in range (1, count + 1):
    x_1 = a + i * width
    x_2 = a + (i - 1) * width

    if (i <= count - 1):
      sum_1 += normal(x_1)

    sum_2 += normal((x_1 + x_2) / 2)
  integral_simpson = width / 3.0 * (1.0 / 2.0 * normal(a) + sum_1 + 2.0 * sum_2 + 1.0 / 2.0 * normal(b))
  return integral_simpson
```

```
def normal (a:float):
  return np.exp(-(1/2) * (m.pow(a, 2)))
```
