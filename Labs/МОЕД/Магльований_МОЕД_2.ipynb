```python
import numpy as np
import math as m
```

```python
def newton_simpson (a:float , b:float, n:int, count:int):
  width = float((b - a) / n)
  x_1 = float(0)
  x_2 = float(0)
  sum_1 = float(0)
  sum_2 = float(0)
  integral_simpson = float(0)

  for i in range (1, n + 1):
    x_1 = a + i * width
    x_2 = a + (i - 1) * width

    if (i <= n - 1):
      sum_1 += simp_equation(x_1, count)

    sum_2 += simp_equation((x_1 + x_2) / 2, count)
  integral_simpson = width / 3.0 * (1.0 / 2.0 * simp_equation(a, count) + sum_1 + 2.0 * sum_2 + 1.0 / 2.0 * simp_equation(b, count))
  return integral_simpson
```

```python
def simp_equation (x:float, count:int):
  simp = m.pow((1 + m.pow(x, 2)/ count), -(count + 1) / 2)
  return simp
```

```python
dfact = lambda n: (n <= 0) or n * dfact(n-2)
```

```python
def student (t:float, count:int):
  b_value = 1 / (2 * m.sqrt(count)) * dfact((2 * count - 1)) / dfact((2 * count))
  s_value = b_value * newton_simpson(0, t, 1000, count)
  return s_value
```

```python
def equation (t:float, count:int, alpha:float):
  equation_value = float(1 - 2 * student(t, count) - alpha)
  return equation_value
```

```python
def half_division_method (a:float, b:float, n:int, precision:float, count:int, alpha:float):
  result = float(0)
  middle = float((b + a) / 2)
  if (n >= 2000):
    print("Найти критическое значение не представляется возможным. Попробуйте изменить значение α.")
    return middle
  else:
    if ( (b - a) <= precision):
      result = middle
    else:
      if (equation(middle, count, alpha) * equation(b, count, alpha) < 0):
        n += 1
        result = half_division_method(middle, b, n, precision, count, alpha)
      else:
        n += 1
        result = half_division_method(a, middle, n, precision, count, alpha)
  return result
```
